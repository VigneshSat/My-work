## 3. Data Preparation



#### Mapping some variables into categorical values for better understanding

```python
# Mapping the values of season column into 1- Spring,2- Summer,3-Fall,4-Winter

Bike['season'] = Bike['season'].map({1:'Spring',2:'Summer',3:'Fall',4:'Winter'})
```

```python
# Check the housing dataframe now

Bike['season'].value_counts()
```

```python
#Mapping the values of month from 1-12 as January to December

Bike['mnth'] = Bike['mnth'].map({1:'January',2:'February',3:'March',4:'April',5:'May',6:'June',7:'July',8:'August',9:'September',10:'October',11:'November',12:'December'})
Bike['mnth'].value_counts()
```

```python
#Mapping the values of weekday from 0-6 as Sunday - Saturday

Bike['weekday'] = Bike['weekday'].map({0:'Sunday',1:'Monday',2:'Tuesday',3:'Wednesday',4:'Thursday',5:'Friday',6:'Saturday'})
Bike['weekday'].value_counts()
```

```python
#Mapping the values of weathers as 1:Clear,2:Mist,3:Light Rain,4:Heavy Rain

Bike['weathersit'] = Bike['weathersit'].map({1:'Clear',2:'Mist',3:'Light Rain',4:'Heavy Rain'})
Bike['weathersit'].value_counts()
```

#### Creating Dummy variables

Here we have some categorical variables like season, weather, month,weekday. To include these in our model, well need to make binary dummy variables.

```python
#Creating dummies for season column as it has 4 seasons

dum_season = pd.get_dummies(Bike['season'])

# Check what the dataset 'dum_season' looks like
dum_season.head()
```

Now, we don't need four columns. We can drop the `Fall` column, as the Fall column can be identified with just the last three columns where  
- `000` will correspond to `Fall`
- `100` will correspond to `Spring`
- `010` will correspond to `Summer`
- `001` will correspond to `Winter`

```python
# Let's drop the first column from dum_season df using 'drop_first = True'
dum_season = pd.get_dummies(Bike['season'], drop_first = True)

# Add the results to the original Bike dataframe
Bike = pd.concat([Bike, dum_season], axis = 1)

# Now let's see the head of our dataframe.
Bike.head()
```

```python
# Let's create dummy for mnth column and drop the first column from dum_months df using 'drop_first = True'
dum_months = pd.get_dummies(Bike['mnth'], drop_first = True)

# Add the results to the original Bike dataframe
Bike = pd.concat([Bike, dum_months], axis = 1)

# Now let's see the head of our dataframe.
Bike.head()
```

```python
# Let's create dummy columns for weekday column and drop the first column from dum_weekday df using 'drop_first = True'
dum_weekday = pd.get_dummies(Bike['weekday'], drop_first = True)

# Add the results to the original Bike dataframe
Bike = pd.concat([Bike, dum_weekday], axis = 1)

# Now let's see the head of our dataframe.
Bike.head()
```

```python
# Let's create dummy columns for weathersit and drop the first column from dum_weather df using 'drop_first = True'
dum_weather = pd.get_dummies(Bike['weathersit'], drop_first = True)

# Add the results to the original Bike dataframe
Bike = pd.concat([Bike, dum_weather], axis = 1)

# Now let's see the head of our dataframe.
Bike.head()
```

```python
#Dropping unnecessary columns

fields_to_drop = ['instant', 'dteday','season','mnth','weekday','weathersit','casual','registered'] 
Bike = Bike.drop(fields_to_drop, axis=1)
Bike.head()
```

## 4. Splitting the Data into Training and Testing Sets

```python
#Importing MinMaxScaler from sklearn library

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

# 5.Building our model

This time, we will be using the **LinearRegression function from SciKit Learn** for its compatibility with RFE (which is a utility from sklearn)

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 15
lm = LinearRegression()
lm.fit(X_train, y_train)

# Running RFE
rfe = RFE(lm, 15)            
rfe = rfe.fit(X_train, y_train)
```

```python
#List of variables which supports RFE and thier ranking

list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
#RFE supported variables

col = X_train.columns[rfe.support_]
col
```

```python
#Unsupported variables

X_train.columns[~rfe.support_]
```

### Building model using statsmodel, for the detailed statistics

```python
# Creating X_test dataframe with RFE selected variables

X_train_rfe = X_train[col]
```

```python
# Adding a constant variable 

import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
# Running the linear model

lm = sm.OLS(y_train,X_train_rfe).fit() 
```

```python
#First regression model

print(lm.summary())
```

#### `Holiday` is insignificant in presence of other variables; can be dropped

```python
X_train_1 = X_train_rfe.drop(["holiday"], axis = 1)
```
